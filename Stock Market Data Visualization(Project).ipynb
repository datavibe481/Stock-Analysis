## Q1 Visualize the Stock Prices using Matplotlib

```python
import matplotlib.pyplot as plt
import yfinance as yf

# Download stock data for the companies
data = yf.download(['MSFT', 'AMZN', 'AAPL', 'GOOG', 'META'], start='2010-01-01')

# Plot the close prices for each company
plt.figure(figsize=(12, 8))
plt.plot(data['Close']['MSFT'], label='Microsoft')
plt.plot(data['Close']['AMZN'], label='Amazon')
plt.plot(data['Close']['AAPL'], label='Apple')
plt.plot(data['Close']['GOOG'], label='Alphabet (Google)')
plt.plot(data['Close']['META'], label='Facebook')
plt.title('Stock Prices of Microsoft, Amazon, Apple, Alphabet (Google), and Facebook')
plt.xlabel('Year')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python

# Plot the close prices for each company
plt.figure(figsize=(12, 8))
plt.plot(data['Close']['MSFT'], label='Microsoft', color='blue', linewidth=2)
plt.plot(data['Close']['AMZN'], label='Amazon', color='red', linewidth=2)
plt.plot(data['Close']['AAPL'], label='Apple', color='green', linewidth=2)
plt.plot(data['Close']['GOOG'], label='Alphabet (Google)', color='purple', linewidth=2)
plt.plot(data['Close']['META'], label='Facebook', color='orange', linewidth=2)
plt.title('Stock Prices of Microsoft, Amazon, Apple, Alphabet (Google), and Facebook', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Close Price', fontsize=15)
plt.legend(fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Q2 Calculate and Visualize the daily Simple rate of return

```python

# Calculate daily simple rate of return
returns = data['Close'].pct_change()

# Plot the daily simple rate of return for each company
plt.figure(figsize=(12, 8))
plt.plot(returns['MSFT'], label='Microsoft', color='blue', linewidth=1)
plt.plot(returns['AMZN'], label='Amazon', color='red', linewidth=1)
plt.plot(returns['AAPL'], label='Apple', color='green', linewidth=1)
plt.plot(returns['GOOG'], label='Alphabet (Google)', color='purple', linewidth=1)
plt.plot(returns['META'], label='Facebook', color='orange', linewidth=1)
plt.title('Daily Simple Rate of Return of Microsoft, Amazon, Apple, Alphabet (Google), and Facebook', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Simple Rate of Return', fontsize=15)
plt.legend(fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Calculate daily simple rate of return
returns = data['Close'].pct_change()

# Plot the daily simple rate of return for each company in a bar graph
plt.figure(figsize=(12, 8))
plt.bar(returns.index, returns['MSFT'], label='Microsoft', color='blue')
plt.bar(returns.index, returns['AMZN'], label='Amazon', color='red')
plt.bar(returns.index, returns['AAPL'], label='Apple', color='green')
plt.bar(returns.index, returns['GOOG'], label='Alphabet (Google)', color='purple')
plt.bar(returns.index, returns['META'], label='Facebook', color='orange')
plt.title('Daily Simple Rate of Return of Microsoft, Amazon, Apple, Alphabet (Google), and Facebook', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Simple Rate of Return', fontsize=15)
plt.legend(fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Q3 Calculate and Visualize the mean rate of return

```python
import numpy as np


# Calculate the daily simple returns of each company
returns = data['Close'].pct_change()

# Calculate the mean returns for each company
mean_returns = returns.mean()

# Plot the mean returns as a bar graph
plt.bar(mean_returns.index, mean_returns.values, color='blue')
plt.title('Mean Rate of Return of Microsoft, Amazon, Apple, Alphabet (Google), and Facebook', fontsize=20)
plt.xlabel('Company', fontsize=15)
plt.ylabel('Mean Rate of Return', fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Q4 Calculate and Visualize the Variances of returns

```python
# Calculate the daily simple returns of each company
returns = data['Close'].pct_change()

# Calculate the variances of returns for each company
variances = returns.var()

# Plot the variances as a bar graph
plt.bar(variances.index, variances.values, color='blue')
plt.title('Variances of Returns for Microsoft, Amazon, Apple, Alphabet (Google), and Facebook', fontsize=20)
plt.xlabel('Company', fontsize=15)
plt.ylabel('Variance of Returns', fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Q5 Calculate and Visualize the Standard Deviation of the returns

```python

# Calculate the daily simple returns of each company
returns = data['Close'].pct_change()

# Calculate the standard deviations of returns for each company
stds = returns.std()

# Plot the standard deviations as a bar graph
plt.bar(stds.index, stds.values, color='blue')
plt.title('Standard Deviations of Returns for Microsoft, Amazon, Apple, Alphabet (Google), and Facebook', fontsize=20)
plt.xlabel('Company', fontsize=15)
plt.ylabel('Standard Deviation of Returns', fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Q6 Correlations between Tech stocks

```python
import pandas as pd

# Calculate the daily simple returns of each company
returns = data['Close'].pct_change()

# Calculate the pairwise correlations of returns
correlations = returns.corr()

# Convert the correlations to a pandas dataframe
correlations_df = pd.DataFrame(correlations)

# Rename the columns and index to reflect the company names
correlations_df.columns = ['Microsoft', 'Amazon', 'Apple', 'Alphabet (Google)', 'Facebook']
correlations_df.index = ['Microsoft', 'Amazon', 'Apple', 'Alphabet (Google)', 'Facebook']

# Display the correlations in a table format
print(correlations_df)
```

```python
import seaborn as sns

# Calculate the daily simple returns of each company
returns = data['Close'].pct_change()

# Calculate the pairwise correlations of returns
correlations = returns.corr()

# Plot the correlations as a heatmap
sns.heatmap(correlations, annot=True, cmap='coolwarm')

# Add labels and title to the plot
plt.title("Correlation of Daily Simple Returns")
plt.xlabel("Companies")
plt.ylabel("Companies")

# Show the plot
plt.show()
```

```python

# Calculate the daily simple returns of each company
returns = data['Close'].pct_change()

# Calculate the mean returns of each company over a 252-day window
mean_returns = returns.rolling(window=252).mean()

# Calculate the pairwise correlations of mean returns
correlations = mean_returns.corr()

# Plot the correlations as a heatmap
sns.heatmap(correlations, annot=True, cmap='coolwarm')

# Add labels and title to the plot
plt.title("Correlation of Mean Returns")
plt.xlabel("Companies")
plt.ylabel("Companies")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
sns.heatmap(returns.corr(),annot=True,cmap='coolwarm')
plt.title('Correlations of Daily Simple Returns')


plt.subplot(2,2,2)
sns.heatmap(mean_returns.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation of Mean Returns')
```

# Thesis Report

#### In the stock market, the correlations between stocks are an important factor to consider for investors and traders. Correlation measures the relationship between two or more stocks and can provide insight into the overall market trend, market risk, and potential diversification opportunities. A positive correlation between stocks indicates that they tend to move in the same direction, while a negative correlation means that they tend to move in opposite directions. Understanding these correlations can help investors make informed decisions about which stocks to invest in and how to allocate their portfolios.
#### One common way to calculate correlations between stocks is by analyzing their daily simple returns or mean returns over a certain period of time. The mean returns are calculated as the average returns over a given window, typically a year or 252 trading days, and provide a clearer picture of the underlying trends in the stock prices. The correlations between the mean returns of different stocks can be visualized using heatmaps, which provide a clear representation of the strength and direction of the relationships.
#### It is important to note that correlations between stocks can change over time and can be affected by various market and economic factors. Hence, it is essential for investors to regularly monitor the correlations between stocks and adjust their portfolios accordingly. By doing so, they can minimize risk and potentially increase returns.

####                                                                           One way to visualize the conclusion from the correlations between stocks is through a scatter plot. A scatter plot shows the relationship between two variables and can help illustrate the correlation between two stocks. For example, if you have calculated the correlations between the mean returns of stocks A and B, you can create a scatter plot with the mean returns of stock A on the x-axis and the mean returns of stock B on the y-axis. Points on the scatter plot closer to a 45-degree line indicate a strong positive correlation, while points closer to a vertical line indicate a strong negative correlation.
#### Another way to visualize the conclusion is through a heatmap. A heatmap is a visual representation of data where values are represented as colors. In the case of stock correlations, a heatmap can display the correlations between multiple stocks, where cells with strong positive correlations are shown in lighter colors and cells with strong negative correlations are shown in darker colors.
#### Both scatter plots and heatmaps can provide a quick and intuitive way to understand the correlations between stocks and can help investors make informed decisions about their portfolios.

####                                                                          In conclusion, the correlations between stocks play a crucial role in the stock market and can provide valuable information to investors and traders. Understanding these correlations can help investors make informed decisions and optimize their portfolios.
